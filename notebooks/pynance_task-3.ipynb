```python
# pynance
import pandas as pd
```

```python
df = pd.read_csv('../data/raw_analyst_ratings.csv')
```

```python
df = df[['date', 'stock']]
```

```python
df["date"] = pd.DataFrame(df["date"])
```

```python
df['date'] = pd.to_datetime(df['date'], utc=True)
```

```python
grouped_data = df.groupby('stock')
```

```python
earliest_dates = grouped_data['date'].min()
latest_dates = grouped_data['date'].max()
```

```python
import yfinance as yf
```

```python
stock_df = pd.DataFrame()
```

```python
for stock, earliest_date in earliest_dates.items():
    latest_date = latest_dates[stock]
    
    stock_data = yf.download(stock, start=earliest_date, end=latest_date)
    
    stock_data['stock'] = stock
    stock_df = pd.concat([stock_df, stock_data], ignore_index=True)
```

```python
stock_df
```

```python
import talib
```

```python
rsi = talib.RSI(stock_df["Close"])
```

```python
rsi
```

```python
macd, signal, hist = talib.MACD(stock_df["Close"], fastperiod=12, slowperiod=26, signalperiod=9)
```

```python
macd, signal, hist
```

```python
stock_df['returns'] = stock_df['Close'].pct_change()
```

```python
stock_df['moving_average'] = stock_df['Close'].rolling(window=3).mean()
```

```python
stock_df['cumulative_returns'] = (1 + stock_df['returns']).cumprod()
```

```python
stock_df['vwap'] = (stock_df['Close'] * stock_df['Volume']).cumsum() / stock_df['Volume'].cumsum()
```
