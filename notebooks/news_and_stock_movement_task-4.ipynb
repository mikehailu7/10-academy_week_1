```python
#news and stock movement
import pandas as pd
import yfinance as yf
```

```python
df = pd.read_csv('../data/raw_analyst_ratings.csv')
```

```python
unique_values = df['stock'].unique()
```

```python
stock_df = pd.DataFrame()
```

```python
for stock in unique_values:
    stock_data = yf.Ticker(stock)
    
    stock_df = pd.concat([stock_df, stock_data.history(period="max")])
```

```python
stock_df
```

```python
stock_df.reset_index(drop=False, inplace=True)
```

```python
stock_df
```

```python
stock_df.rename(columns={'Date': 'date'}, inplace=True)
```

```python
df["date"] = pd.DataFrame(df["date"])
```

```python
df['date'] = pd.to_datetime(df['date'], utc=True)
```

```python
merged_df = pd.merge(df, stock_df, on='date', how='inner')
```

```python
merged_df
```

```python
import nltk
from textblob import TextBlob
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
df
```

```python
for headline in df["headline"].head(20):
    blob = TextBlob(headline)
    sentiment = blob.sentiment.polarity
    df["sentiment"] = sentiment

    if sentiment > 0:
        sentiment_label = "Positive"
    elif sentiment < 0:
        sentiment_label = "Negative"
    else:
        sentiment_label = "Neutral"

    print(f"Headline: {headline}")
    print(f"Sentiment: {sentiment_label}")
    print(f"Headline: {df['sentiment']}")
```

```python
df.sort_values('date', inplace=True)

df['return'] = stock_df['Close'].pct_change() * 100
```

```python
df
```

```python
from scipy.stats import pearsonr, spearmanr
```

```python
daily_sentiments = df.groupby('date')['sentiment'].mean()
```

```python
daily_returns = df.groupby('date')['return'].mean()
```

```python
df_2  = pd.merge(daily_sentiments, daily_returns, on='date', how='inner')
```

```python
spearman_corr, spearman_pvalue = spearmanr(df['sentiment'], df['return'])
print(f"Spearman correlation coefficient: {spearman_corr}")
print(f"P-value: {spearman_pvalue}")
```
